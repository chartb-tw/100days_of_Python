```python
import bs4
import requests
import json
import csv
```

```python
resp = requests.get("https://sckans.edu/~sireland/radio/code.html")
resp.raise_for_status()
soup = bs4.BeautifulSoup(resp.text,"html.parser")
soup.prettify()
```

```python
# If the above doesn't work, could use the local copy:
with open("Morse Code and Phonetic Alphabet Page.html") as file:
    content = file.read()
local_soup = bs4.BeautifulSoup(content, 'html.parser')
local_soup.prettify()
```

```python
soup.find_all(name="table")
```

```python
soup.find_all(name="table")[0].find_all(name = "tr")
```

```python
pa_list = soup.find_all(name="table")[0].find_all(name = "tr")[1:]
pa_dict = {}

for alph in pa_list:
    pa_dict[alph.find_all(name = "td")[0].getText().strip()]=alph.find_all(name = "td")[1].getText().strip()
```

```python
pa_numlist = soup.find_all(name="table")[1].find_all(name = "tr")[1:]
for num in pa_numlist:
    pa_dict[num.find_all(name = "td")[0].getText().strip()] = num.find_all(name = "td")[1].getText().strip()
```

Could convert the dictionary into a [JSON object from the `json` package](https://docs.python.org/3/library/json.html#basic-usage):

```python
with open("morse_code.json",'w') as file:
    json.dump(pa_dict, file)
```

Could also create a `.csv` file, using the [`csv` package](https://docs.python.org/3/library/csv.html) and looping through the [`pa_dict` by element and value](https://www.w3schools.com/python/python_dictionaries_loop.asp):

```python
with open('morse_code.csv','w',newline='') as csv_file:
    csv_headers = ['character','morse code']
    mc_csv_wt = csv.DictWriter(csv_file, fieldnames=csv_headers)
    mc_csv_wt.writeheader()
    for el in pa_dict:
        mc_csv_wt.writerow({csv_headers[0]:el, csv_headers[1]:pa_dict[el]})
```

```python
morse_dict = pa_dict.copy() # could set this to be sourced from any other file
usr_req = input("Give a word to convert into Morse Code: ").upper()
out = []
for char in usr_req:
	if char in morse_dict.keys():
		out.append(morse_dict[char])
print(out)
```
