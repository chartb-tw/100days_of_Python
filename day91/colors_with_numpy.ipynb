```python
# via https://www.codespeedy.com/convert-rgb-to-hex-color-code-in-python/
def rgb_to_hex(rgb):
    r,g,b = rgb[:3]
    return '#%02x%02x%02x' % (r,g,b)
```

```python
def col_dist(rgb1, rgb2):
    r1,g1,b1 = rgb1[:3]
    r2,g2,b2 = rgb2[:3]
    dis = pow(abs(r1 - r2),2) + pow(abs(g1 - g2),2) + pow(abs(b1 - b2),2)
    return math.sqrt(dis)
```
